```python
# Importing the needed Libaries 

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Reading the csv file 

nba = pd.read_csv("nba_2013.csv")
nba
```

```python
# To view all the columns

pd.set_option('display.max_columns', None)
```

```python
# To view the types of data

nba.dtypes
```

```python
# Selecting data types that are objects

nba.select_dtypes(include=["object"]) 
```

```python
# Selecting data that are number data types

nba.select_dtypes(include= ["number"])
```

```python
# Creating new columns

nba["pts_per_game"] = nba["pts"] / nba["g"]
nba["ast_per_game"] = nba["ast"] / nba["g"]
nba["total_rebound_per_game"] = nba["trb"] / nba['g']
nba
```

```python
# Who are the top 5 scorer per game during the 2013 season?

top_5_scorer = nba.groupby("player")["pts_per_game"].mean().reset_index().sort_values(by= "pts_per_game", ascending=False)
top_5_scorer.head(5)
```

```python
# Which player scored the most points during the 2013 season?

leading_scorer =nba[nba["pts_per_game"]==nba['pts_per_game'].max()]
leading_scorer
```

```python
# Which team did the leading scorer play for? 

leading_scorer.bref_team_id
```

```python
# Which positions is leading scorer playing as? 

leading_scorer.pos
```

```python
# How many games did the leading scorer play?

leading_scorer.g
```

```python
# Which players scored more than 25 points per game?

scored_more_than_25 = nba[nba["pts_per_game"] > 25]
scored_more_than_25
```

```python
# Which player had the most assist per game?

nba[nba["ast_per_game"]==nba["ast_per_game"].max()]
```

```python
# Which player averaged a double double during the 2013 season.

nba[(nba["pts_per_game"]>10) & (nba["ast_per_game"]>10)]
```

```python
# Using Groupby() function, group the players that played for the Clippers during the 2013 season?

Clippers = nba.groupby('bref_team_id').get_group("LAC")
Clippers
```

```python
# Who were the point guard in the Clippers team?

Point_guard = ["PG"]
Clippers[Clippers["pos"].isin(Point_guard)]
```

```python
# Which players in the Clippers are younger than 28 years old?

Clippers[Clippers["age"]<28]
```

```python
# Which player scored the most points per game in the Clippers?

Clippers[Clippers["pts_per_game"]==Clippers.pts_per_game.max()]
```

```python
# Which positions scored the most points?

nba.groupby("pos").describe()
```

```python
# From the previous analysis, filter to only view the 5 tradition positions?

position = ["C", "PF", "SF", "SG", "PG"]

Tradition_position = nba[nba["pos"].isin(position)]
Tradition_position
```

```python
Tradition_position.groupby("pos").pts_per_game.describe()
```

```python
matrix = Tradition_position.corr()
matrix
```

```python
unstack_matrix=matrix.unstack()
unstack_matrix
```

```python
sorted_pairs = unstack_matrix.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>.5]
high_corr
```
